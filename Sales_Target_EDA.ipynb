# Data Anlysis & Visualization

## Sales Over Time

```python
# Concatenate sales data for all years
all_sales = pd.concat([s_2017, s_2018, s_2019], ignore_index=True)
```

```python
all_sales.head()
```

```python
all_sales.shape
```

```python
all_sales.tail()
```

```python
# Merge sales data with dates data based on Issue Date
merged_sales = pd.merge(all_sales, dates_table, left_on='Issue Date', right_on='Dates')
```

```python
merged_sales.head()
```

```python
# Group sales data by Date and sum the Sales column
sales_over_time = merged_sales.groupby('Issue Date')['Sales'].sum()
```

```python
sales_over_time
```

```python
# Convert index to datetime
sales_over_time.index = pd.to_datetime(sales_over_time.index)
```

```python
sales_over_time.info()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(sales_over_time.index, sales_over_time.values, marker='o', linestyle='-', color='b')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Sales by years

# Group sales data by Year and sum the Sales column
sales_by_year = merged_sales.groupby('Year')['Sales'].sum()
```

```python
sales_by_year
```

```python
# Plotting sales over year using a bar graph
plt.figure(figsize=(10, 6))
sales_by_year.plot(kind='bar', color='b')
plt.title('Sales Over Time by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Sales by Seasonal Pattern

To analyze seasonal patterns, we can use techniques like moving averages or decomposition to identify recurring patterns in sales data

```python
# Calculate 30-day moving average to smooth out noise
sales_by_date_moving_avg = sales_over_time.rolling(window=30).mean()
sales_by_date_moving_avg
```

```python
plt.figure(figsize=(12, 6))
plt.plot(sales_over_time, label='Daily Sales')
plt.plot(sales_by_date_moving_avg, label='30-Day Moving Avg', color='orange')
plt.title('Overall Sales Trend with 30-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ManagerWise Analysis

# Merge sales data with SalesPerson data based on SalesPerson ID
merged_sales_with_SalesPerson = pd.merge(all_sales, salesperson, left_on='SalesPerson ID', right_on='SalesPerson ID')

merged_sales_with_SalesPerson.head()
```

```python
# Group sales data by Manager and sum the Sales column
sales_by_manager = merged_sales_with_SalesPerson.groupby('Manager')['Sales'].sum()

sales_by_manager.head()
```

```python
# Create a pie chart for manager-wise analysis
plt.figure(figsize=(10, 6))
plt.pie(sales_by_manager, labels=sales_by_manager.index, autopct='%.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Manager-wise Analysis')
#plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
# SalesPersonWise Analysis
# Group sales data by SalesPerson and sum the Sales column
sales_by_SalesPerson = merged_sales_with_SalesPerson.groupby('SalesPerson')['Sales'].sum()

sales_by_SalesPerson
```

```python
# Sort SalesPerson by sales
sorted_sales_by_SalesPerson = sales_by_SalesPerson.sort_values(ascending=False)

# Plotting manager-wise analysis using a bar graph
plt.figure(figsize=(10, 6))
ax = sorted_sales_by_SalesPerson.plot(kind='bar', color='b')
plt.title('SalesPerson-wise Analysis')
plt.xlabel('SalesPerson')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)

# Annotate each bar with its sales value
for i, v in enumerate(sorted_sales_by_SalesPerson):
    ax.text(i, v + 5000, '${:,.2f}'.format(v), ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Group sales data by different hierarchy levels and sum the Sales column
sales_by_salesperson = merged_sales_with_SalesPerson.groupby('SalesPerson')['Sales'].sum()
sales_by_supervisor = merged_sales_with_SalesPerson.groupby('Supervisor')['Sales'].sum()
sales_by_manager = merged_sales_with_SalesPerson.groupby('Manager')['Sales'].sum()
sales_by_team = merged_sales_with_SalesPerson.groupby('Team')['Sales'].sum()
```

```python
# Visualize sales performance at different hierarchy levels
plt.figure(figsize=(12, 8))
plt.barh(sales_by_salesperson.index, sales_by_salesperson.values, color='blue', label='Salesperson')
plt.barh(sales_by_supervisor.index, sales_by_supervisor.values, color='green', label='Supervisor')
plt.barh(sales_by_manager.index, sales_by_manager.values, color='orange', label='Manager')
plt.barh(sales_by_team.index, sales_by_team.values, color='red', label='Team')
plt.xlabel('Total Sales')
plt.ylabel('Hierarchy Level')
plt.title('Sales Performance by Hierarchy Level')
plt.legend()
plt.tight_layout()
plt.show()
```
